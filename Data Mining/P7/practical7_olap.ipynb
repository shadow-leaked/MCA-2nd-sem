# OLAP Operations

```python
import pandas as pd
```

```python
data = {
    'Region': ['East','East','West','West'],
    'Year': [2024,2025,2024,2025],
    'Sales': [100,200,300,400]
}
```

```python
df = pd.DataFrame(data)

print(df)
```

```python
print("Roll-Up Operation")
print(df.groupby('Region')['Sales'].sum())
```

```python
print("Slice Operation")
print(df[df['Region'] == 'East'])
```

```python
print("Dice Operation")
print(df[(df['Region'] == 'West') & (df['Year'] == 2025)])
```

## Conclusion

OLAP operations were successfully performed on sales data.
