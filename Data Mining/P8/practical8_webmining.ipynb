# Web Mining Basics

```python
from collections import Counter
```

```python
with open("weblog.txt", "r") as file:
    data = file.readlines()
```

```python
paths = [line.strip() for line in data]

print(paths)
```

```python
count = Counter(paths)
```

```python
print("Page Visit Frequency")
print(count)
```

## Conclusion

Frequent user page access patterns were successfully analyzed from web logs.
