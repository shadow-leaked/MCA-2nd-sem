# ETL Process Implementation

## Aim
To perform Extract, Transform and Load (ETL) process using Python and SQL.

```python
import pandas as pd
import sqlite3
```

```python
csv_file = "s_data.csv"
```

import pandas as pd

```python
df = pd.read_csv(csv_file)

print("===== EXTRACTED DATA =====")
print(df)
```

```python
df = df.dropna()
```

```python
if "Marks" in df.columns:
    df["Marks"] = df["Marks"] + 5
```

```python
df = df.drop_duplicates()
```

```python
print("\n===== TRANSFORMED DATA =====")
print(df)
```

```python
conn = sqlite3.connect("student.db")
```

```python
df.to_sql(
    "students",          # table name
    conn,
    if_exists="replace", # replace old table
    index=False
)
```

```python
print("Data loaded into database successfully")
```

```python
query = "SELECT * FROM students"

result = pd.read_sql(query, conn)

print("\n===== DATA FROM DATABASE =====")
print(result)
```

```python
conn.close()
```
