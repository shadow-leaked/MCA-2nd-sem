## Aim
To perform Extract, Transform and Load (ETL) process using Python and SQL.

```python
import pandas as pd
import sqlite3
```

import pandas as pd

```python
data = {
    "Name": ["Aditya", "Rajjit", "Nelson"],
    "Marks": [70, 85, 90]
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Extracted Data")
print(df)
```

```python
df["Marks"] = df["Marks"] + 5
```

```python
print("Transformed Data")
print(df)
```

```python
conn = sqlite3.connect("student.db")
```

```python
df.to_sql("students", conn, if_exists="replace", index=False)
```

```python
print("Data loaded into database successfully")
```

```python
result = pd.read_sql("SELECT * FROM students", conn)
print(result)
```
